```python
while True:
    try:
        load_more_comments = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "svg [aria-label='Load more comments']")))
        load_more_comments.click()
        time.sleep(30)
    except (NoSuchElementException, TimeoutException) as e:
        break    
```

```python
# Get the initial scroll height
last_height = driver.execute_script("return arguments[0].scrollHeight", scroll)

while True:
    try:
        # Scroll to the bottom of the div using JavaScript
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll)
        time.sleep(8)

        # Get the new scroll height after scrolling
        new_height = driver.execute_script("return arguments[0].scrollHeight", scroll)

        # Check if we've reached the bottom of the div
        if new_height == last_height:
            break  # Exit the loop if no new content is loaded

        last_height = new_height  # Update the last scroll height
    except:
        break
```

```python
# Find all div elements inside the comment section
comment_section = driver.find_element(By.XPATH, '/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/section/main/div/div[1]/div/div[2]/div/div[2]/div/div[2]')
comment_divs = comment_section.find_elements(By.XPATH, './div')
```

```python
# Store the div elements in a list
div_elements_list = []

for div_element in comment_divs:
    div_elements_list.append(div_element)
```

```python
print(div_elements_list[0:5])
```

```python
# Iterate through each comment element
comment_list = []
for comment in comment_divs:
    try:
        # Extract username
        # username_element = comment.find_element(By.XPATH, './div[@class="username"]')
        # username = username_element.text.strip()  # Assuming the username is text inside the username element
        # Extract comment text
        comment_element = comment.find_element(By.XPATH, './div[1]/div/div[2]/div[1]/div[1]/div/div[2]/span')
        comment_text = comment_element.text.strip()  # Assuming the comment text is text inside the comment element

        # Print or store the username and comment text
        # print("Username:", username)
        print("Comment:", comment_text)
        comment_list.append(comment_text)
    except NoSuchElementException as e:
        # Handle the case where username or comment text element is not found
        print("Error:", e)
        # Optionally, you can continue to the next comment or take appropriate action
        continue
```

```python
# Open a CSV file in write mode
with open('comments.csv', 'w', newline='', encoding='utf-8') as csvfile:
    # Create a CSV writer object
    csv_writer = csv.writer(csvfile)
    
    # Write header row
    csv_writer.writerow(['COMMENTS'])
    
    # Write data rows
    for comments in comment_list:
        csv_writer.writerow([comments])
```
